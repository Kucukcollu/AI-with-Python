### Subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,"blue")

plt.subplot(2,2,2)
plt.plot(y,x,"yellow")

plt.subplot(2,2,3)
plt.plot(y,x,"red")

plt.subplot(2,2,4)
plt.plot(x,x**2,"black")

plt.show() # if we use in here that will not cause any problem
```

### Usage with matplotlib object

```python
x = np.arange(1,6)
y = np.arange(2,11,2)

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.2,0.3])

axes1.plot(x,y)
axes1.set_xlabel("Outer X")
axes1.set_ylabel("Outer Y")
axes1.set_title("Outer Graph")

axes2.plot(x,y)
axes2.set_xlabel("Inner X")
axes2.set_ylabel("Inner Y")
axes2.set_title("Inner Graph")

plt.show()
```

```python
x
```

```python
y
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

for a in axes:
    a.plot(x,y)

plt.tight_layout() # to see more clear
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1)

axes[0].plot(x,y)
axes[0].set_title("First axes")
axes[1].plot(x,x**2)
axes[1].set_title("Second axes")

plt.tight_layout() # to see more clear
```

### Adjust the dimension of the axes

```python
fig = plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple")
```

```python
fig = plt.figure(figsize=(15,5))

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color="purple")
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,6))

axes[0].plot(x,y,"red")
axes[0].set_title("first graph")

axes[1].plot(x,x**4,"blue")
axes[1].set_title("second graph")

plt.tight_layout()
```

### Configure plot

```python
fig=plt.figure(figsize=(6,4))

axes=fig.add_axes([0,0,1,1])

axes.plot(x,x**0.5,color="purple",label="square root")
axes.plot(x,x**2,color="red",label="square")
axes.plot(x,x**3,color="yellow",label="cube")
axes.plot(x,y,color="#31E818",label="y") # hex color code

axes.legend()
```

```python
figure=plt.figure()

axes=figure.add_axes([0,0,1,1])

# axes.plot(x,x**3,color="red",linewidth=3,linestyle="--") # or we can use also lw instead of linewidth

axes.plot(x,x**2,color="black",lw=3,linestyle="-.",marker="o",markersize=15,markerfacecolor="yellow",
          markeredgecolor="purple",markeredgewidth=4)

# set the desired limits
axes.set_xlim(0,10)
axes.set_ylim(0,30)
```
